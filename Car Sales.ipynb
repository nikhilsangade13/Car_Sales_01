```python
#Analysis of Volkswagen Car Sales

vw=df.groupby('Company').get_group('Volkswagen')
vw=vw.reset_index()
vw
```

```python
vw= vw[['Date','Model']]
vw
```

```python
# Convert the 'Date' column to datetime format
vw.loc[:, 'Date'] = pd.to_datetime(vw['Date'])
vw.head(2)
```

```python
vw['Date'] = pd.to_datetime(vw['Date'])
vw.loc[:, 'Month'] = vw['Date'].dt.month
vw.loc[:, 'Year'] = vw['Date'].dt.year
vw.head(2)
```

```python
vw= vw[['Model','Month']]
vw
```

```python
#monthly car sales of Volkswagen

sns.displot(data=(vw[['Month']]))
```
